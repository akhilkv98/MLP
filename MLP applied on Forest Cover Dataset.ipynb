```

import numpy as np


from sklearn.datasets import fetch_covtype

from sklearn.neural_network import MLPClassifier


X, y = fetch_covtype(return_X_y=True)
 

print(type(X))


print(X.shape)

print(y.shape)


X_10 = X[:,:10]


print(X_10.shape)


from sklearn.model_selection import train_test_split

X10_train, X10_test, y10_train, y10_test = train_test_split(X_10, y, test_size=0.1,
stratify=y, random_state=42)


print(X10_test.shape)




from sklearn.model_selection import train_test_split
 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
stratify=y, random_state=42)


from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train) 
X_test_std = sc.transform(X_test)


mlpClassifier = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(50, 25), random_state=1)


mlpClassifier.fit(X_train_std, y_train)

score = mlpClassifier.score(X_test_std, y_test)

print(score)
```
